```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.get_dataset_names()
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.set_style('darkgrid')
```

```python
sns.displot(data=iris,x='species',height=4)
```

```python
sns.displot(data=iris,x='petal_width',height=4)
```

```python
sns.displot(data=iris,x='sepal_length',y='sepal_width',hue='species',palette='rainbow_r')
```

```python
sns.displot(data=iris,x='sepal_length',y='sepal_width',hue='species',palette='rainbow_r',kind='kde')
```

```python
sns.distplot(iris['petal_length'],color='g')
```

```python
sns.distplot(iris['petal_length'],color='g',bins=10)
```

```python
sns.jointplot(data=iris,x='sepal_length',y='sepal_width',hue='species',palette='rainbow_r')
```

```python
sns.jointplot(data=iris,x='petal_length',y='petal_width',hue='species',palette='rainbow_r')
```

```python
sns.pairplot(iris,hue='species')
```

```python
sns.rugplot(data=iris,x='sepal_length',hue='species',height=0.5)
```

```python
sns.rugplot(data=iris,x='petal_length',hue='species',height=0.5)
```

```python
sns.barplot(x="petal_length",y="petal_width",data=iris,hue='species')
```

```python
iris.groupby('species').mean()['sepal_length']
sns.barplot(x="sepal_length",y="sepal_width",data=iris,hue='species',estimator=np.std)
```

```python
sns.countplot(x='sepal_length',data=iris)
```

```python
sns.countplot(x='sepal_width',hue='species',data=iris)
```

```python
sns.boxplot(y='petal_length',data=iris)
```

```python
sns.boxplot(x='species',y='petal_length',data=iris,hue='species')
```

```python
sns.violinplot(x='species',y='sepal_width',data=iris,hue='species')
```

```python
sns.violinplot(x='species',y='sepal_width',data=iris,hue='species',scale_hue=False)
```

```python
sns.stripplot(x='petal_length',y='species',data=iris)
```

```python
sns.stripplot(x='species',y='sepal_width',data=iris,hue='species')
```

```python
sns.stripplot(x='species',y='sepal_length',data=iris,hue='species',jitter=False)
```

```python
sns.swarmplot(x='species',y='sepal_length',data=iris)
```

```python
sns.swarmplot(x='species',y='petal_length',data=iris,hue='species',size=10)
```

```python
sns.factorplot(data=iris,x='species',y='sepal_width')
```

```python
sns.factorplot(data=iris,x='species',y='sepal_width',kind='bar')
```

```python
sns.FacetGrid(iris,hue='species',size=5)\
    .map(plt.plot,'sepal_length')\
    .add_legend();
```

```python
sns.FacetGrid(iris,hue='species',size=5)\
    .map(plt.scatter,'petal_length','sepal_length')\
    .add_legend();
```

```python
sns.FacetGrid(iris,hue='species',size=5)\
    .map(plt.plot,'petal_length','petal_width')\
    .add_legend();
```

```python
sns.lmplot(x='sepal_length',y='sepal_width',data=iris,hue='species',size=5)
```

```python
sns.lmplot(x='sepal_length',y='petal_width',hue='species',data=iris,size=5,col='species')
```

```python
sns.lmplot(x='sepal_length',y='petal_width',hue='species',data=iris,size=5,col='species',markers=['.','*','o'])
```

```python
sns.heatmap(iris.corr(),square=True)
```
